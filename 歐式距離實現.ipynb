# 歐氏距離(Euclidean Distance)

原文網址：https://kknews.cc/code/v3laxal.html

```
# -*- coding: utf-8 -*- 
from numpy import * 
vector1 = mat([1,2,3]) 
vector2 = mat([4,5,6]) 
print(sqrt((vector1-vector2)*((vector1-vector2).T))) 

import numpy as np 
x = np.random.random(128) 
y = np.random.random(128) 
# solution1 
dist1 = np.linalg.norm(x - y) 

# solution2 
dist2 = np.sqrt(np.sum(np.square(x - y))) 
print('x: ', x) 
print('y: ', y) 

# solution3 
from scipy.spatial.distance import pdist 
X = np.vstack([x,y]) 
dist3 = pdist(X)[0] 

print('solution1 dist:', dist1) 
print('solution2 dist:', dist2) 
print('solution3 dist:', dist3)
```

# 標準化歐氏距離 (Standardized Euclidean distance )

原文網址：https://kknews.cc/code/v3laxal.html

```
# -*- coding: utf-8 -*- 
import numpy as np 
x = np.random.random(128) 
y = np.random.random(128) 
X = np.vstack([x, y]) 
print("x:", x)
print("y:", y)
print("X:", X)

#方法一：根據公式求解 
sk = np.var(X, axis=0, ddof=1) 
d1 = np.sqrt(((x - y) ** 2 /sk).sum()) 
print('d1:', d1) 

#方法二：根據scipy庫求解 
from scipy.spatial.distance import pdist 
d2 = pdist(X, 'seuclidean')[0] 
print('d2:', d2)
```
