# 讀取影像
※使用plt.imshow()要記得使用RGB格式，不然顏色會很奇怪。

## 下載face.jpg

```
!wget https://pbs.twimg.com/media/ECtgOCaUEAAFAMI.jpg -O face.jpg
```

```
import cv2
import matplotlib.pyplot as plt


plt.figure(figsize=(15, 15))

img = cv2.cvtColor(cv2.imread("face.jpg"), cv2.COLOR_BGR2RGB)
plt.subplot(321)
plt.title("Original image")
plt.imshow(img)

gray_img = cv2.cvtColor(cv2.imread("face.jpg", cv2.IMREAD_GRAYSCALE), cv2.COLOR_BGR2RGB)
plt.subplot(322)
plt.title("Gray image")
plt.imshow(gray_img)

gdal_img = cv2.cvtColor(cv2.imread("face.jpg", cv2.IMREAD_LOAD_GDAL), cv2.COLOR_BGR2RGB)
plt.subplot(323)
plt.title("GDAL image")
plt.imshow(gdal_img)

# 將彩色影像尺寸縮成1/2
reduce_1_divide_by_2_img = cv2.cvtColor(cv2.imread("face.jpg", cv2.IMREAD_REDUCED_COLOR_2), cv2.COLOR_BGR2RGB)
plt.subplot(324)
plt.title("Reduce size to 1/2 color image")
plt.imshow(reduce_1_divide_by_2_img)

# 將灰階影像尺寸縮成1/2
reduce_1_divide_by_2_gray_img = cv2.cvtColor(cv2.imread("face.jpg", cv2.IMREAD_REDUCED_GRAYSCALE_2), cv2.COLOR_BGR2RGB)
plt.subplot(325)
plt.title("Reduce size to 1/2 gray image")
plt.imshow(reduce_1_divide_by_2_gray_img)

plt.show()
```

# 顯示影像

```
import cv2
import numpy as np

img = np.ones((640, 480, 3)) * 255
cv2.imshow("Image", img)
key = cv2.waitKey(0)

if key == ord('q'):
    cv2.destroyWindow("Image")
    # cv2.destroyAllWindows()
```
