# 取得影像屬性

```python
import cv2

img = cv2.imread("face.jpg")
print("Image shape: ", img.shape)
print("Image size pixels: ", img.size)
print("Image data type: ", img.dtype)
```

# 取得與更改像素值<br>item(h, w, channel)、itemset((h, y, channel), value)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.cvtColor(cv2.imread("face.jpg"), cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))

# 取得 h、w、channel索引值為1的像素值
print(img.item(1, 1, 1))
plt.subplot(121)
plt.imshow(img)

# 修改 h、w、channel索引值為1的像素值成255
img.itemset((1,1,1), 0)
print(img.item(1, 1, 1))
plt.subplot(122)
plt.imshow(img)

plt.show()
```

# 通道拆分-cv2.split(image)

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.cvtColor(cv2.imread("face.jpg"), cv2.COLOR_BGR2RGB)
b, g, r = cv2.split(img)

plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.title("Original img")
plt.imshow(img)

plt.subplot(222)
plt.title("Blue channel")
plt.imshow(b, cmap="gray")

plt.subplot(223)
plt.title("Green channel")
plt.imshow(g, cmap="gray")

plt.subplot(224)
plt.title("Red channel")
plt.imshow(r, cmap="gray")

plt.show()
```

# 通道合併

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("face.jpg")
b, g, r = cv2.split(img)

RGB_image = cv2.merge([r,g,b])

plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.title("Original img")
plt.imshow(img)

plt.subplot(122)
plt.title("Swap red and blue channel")
plt.imshow(RGB_image)

plt.show()
```
