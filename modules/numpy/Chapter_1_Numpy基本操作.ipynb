# 產生陣列

```python
import numpy as np

array = np.arange(0, 100)
print(array)
```

# 產生隨機整數

```python
import numpy as np

random_int_array = np.random.randint(0, 10000, size=100)
print(random_int_array)
```

# 產生隨機亂數

```python
import numpy as np

random_float_array = np.random.random(size=100)
print(random_float_array)
```

# 清單轉Numpy陣列

```python
import numpy as np

a = [x for x in range(100)]
print(type(a))
print(a)
print()

a = np.array(a)
print(type(a))
print(a)
```
